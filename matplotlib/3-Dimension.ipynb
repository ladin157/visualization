# Three-Dimensional Plotting

Three-dimensional plots are enabled by importing the __mplot3d toolkit__, included with the Matplotlib package.

Tutorial [here](https://matplotlib.org/tutorials/toolkits/mplot3d.html#sphx-glr-tutorials-toolkits-mplot3d-py).

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from matplotlib import cm

%matplotlib inline
```

## 1. Line Plot

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
z = np.linspace(0,1,100)
x = z*np.sin(20*z)
y = z*np.cos(20*z)
ax.plot3D(x,y,z, 'gray')
ax.set_title('3D line plot')
```

## 2. Scatter Plot

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
z = np.linspace(0, 1, 100)
x = z * np.sin(20 * z)
y = z * np.cos(20 * z)
c = x+y
ax.scatter(x,y,z,c=c)
ax.set_title('3d Scatter plot')
```

## 3. 3D Contour Plot

```python
def f(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))

def g(x,y):
    return x**2 + y**2

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)
# print(x)
# print(y)

X, Y = np.meshgrid(x, y)
# print(X)
# print(Y)
Z = g(X, Y)

# print(Z)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap=cm.coolwarm)
```

## 4. Wireframe

```python
def f(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))

def g(x,y):
    return x**2 + y**2

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)
# print(x)
# print(y)

X, Y = np.meshgrid(x, y)
# print(X)
# print(Y)
Z = f(X, Y)

# print(Z)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_wireframe(X, Y, Z, cmap=cm.cool)
```

## 5. 3D Surface

```python
def f(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))

def g(x,y):
    return x**2 + y**2

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)
# print(x)
# print(y)

X, Y = np.meshgrid(x, y)
# print(X)
# print(Y)
Z = f(X, Y)

# print(Z)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, cmap=cm.viridis)
```

## 6. 3D Bar

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
for c, z in zip(['r', 'g', 'b', 'y'], [30, 20, 10, 0]):
    xs = np.arange(20)
    ys = np.random.rand(20)

    # You can provide either a single color or an array. To demonstrate this,
    # the first bar of each set will be colored cyan.
    cs = [c] * len(xs)
    cs[0] = 'c'
    ax.bar(xs, ys, zs=z, zdir='y', color=cs, alpha=0.8)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
```

## 7. 3D Subplots

```python
# set up a figure twice as wide as it is tall
fig = plt.figure(figsize=plt.figaspect(0.5))

#===============
#  First subplot
#===============
# set up the axes for the first plot
ax = fig.add_subplot(1, 2, 1, projection='3d')

# plot a 3D surface like in the example mplot3d/surface3d_demo
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)
surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
ax.set_zlim(-1.01, 1.01)
fig.colorbar(surf, shrink=0.5, aspect=10)

#===============
# Second subplot
#===============
# set up the axes for the second plot
ax = fig.add_subplot(1, 2, 2, projection='3d')

# plot a 3D wireframe like in the example mplot3d/wire3d_demo
ax.plot_wireframe(X, Y, Z, rstride=10, cstride=10)
```

## 8. Text

```python
r'$\alpha_i> \beta_i$'

t = np.arange(0.0, 2.0, 0.01)
s = np.sin(2*np.pi*t)

plt.plot(t,s)
plt.title(r'$\alpha_i> \beta_i$', fontsize=20)

plt.text(0.6, 0.6, r'$\mathcal{A}\mathrm{sin}(2 \omega t)$', fontsize = 20)
plt.text(0.1, -0.5, r'$\sqrt{2}$', fontsize=10)
plt.xlabel('time (s)')
plt.ylabel('volts (mV)')
```

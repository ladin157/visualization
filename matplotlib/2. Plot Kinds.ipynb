## 1. Bar Plot

A bar chart of bar graph is a chart or graph that presents categorical data with rectangular bars with heights or lengths proportional to the values that they present.

Matplotlib API provides the **bar()** function that can be used in the MATLAB style use as well as object oriented API. 

`ax.bar(x, height, width, bottom, align)`

x	| sequence of scalars representing the x coordinates of the bars. align controls if x is the bar center (default) or left edge.

height |	scalar or sequence of scalars representing the height(s) of the bars.

width |	scalar or array-like, optional. the width(s) of the bars default 0.8

bottom |	scalar or array-like, optional. the y coordinate(s) of the bars default None.

align |	{‘center’, ‘edge’}, optional, default ‘center’

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
langs = ['C', 'C++','Java', 'Python', 'PHP']
students = [23, 17, 35, 29, 12]
ax.bar(x=langs, height=students)
```

```python
data = [[30, 25, 50, 20],
        [40, 23, 51, 17],
        [35, 22, 45, 19]]
X = np.arange(4)
print(X)
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.bar(X + 0.00, data[0], color='b', width=0.25)
ax.bar(X + 0.25, data[1], color='g', width=0.25)
ax.bar(X + 0.5, data[2], color='r', width=0.25)
ax.legend(labels=('CS','IT','E&TC'))
```

```python
N = 5
menMeans = (20, 35, 30, 35, 27)
womenMeans = (25, 32, 34, 20, 25)
ind = np.arange(N) # the x locations for the groups
print(ind)
width = 0.35
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(ind, menMeans, width, color='r')
ax.bar(ind, womenMeans, width, bottom=menMeans, color='b')
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(ind, ('G1', 'G2', 'G3', 'G4', 'G5'))
ax.set_yticks(np.arange(0, 81, 10))
ax.legend(labels=['Men', 'Women'])
```

## 3. Pie Chart

A Pie chart only display one series of data. Pie charts show the size of items (called wedge) in one data series, propotional to the sum of the items. The data points in a pie chart are shown as a percentage of the whole pie.

Matplotlib API has a **pie()** functiion that generates a pie diagram representing data in an array.

**Parameters**

**Following table lists down the parameters foe a pie chart −**

----------------------------------------------------------------------------------
 x	| array-like. The wedge sizes.                                               

labels |	list. A sequence of strings providing the labels for each wedge.

Colors |	A sequence of matplotlibcolorargs through which the pie chart will cycle. If None, will use the colors in the currently active cycle.

Autopct |	string, used to label the wedges with their numeric value. The label will be placed inside the wedge. The format string will be fmt%pct.

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.axis('equal')
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23, 25, 35, 12, 20]
ax.pie(students, labels=langs, autopct='%1.2f%%')
```

## 4. Scatter Plot

Scatter plots are used to plot data points on horizontal and vertical axis in the attempt to show how much one variable is affected by another. Each row in the data table is represented by a marker the position depends on its values in the columns set on the X and Y axes. A third variable can be set to correspond to the color or size of the markers, thus adding yet another dimensions to the plot.

```python
x1 = np.random.rand(20)
print(x1)
x2 = np.random.rand(20)+5
print(x2)
x = sorted(np.random.randint(20, size=20))
print(x)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.scatter(x, x1)
ax.scatter(x, x2, color='r')
ax.set_title("Scatter Plot")
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend(labels=['x1','x2'])
```

## 5. Contour Plot

Contour plots (sometime called Level Plots) are a way to show a three-dimensional surface on a two-dimensional plane. It graphs two predictor variables X Y on the y-axis and a response variable Z as contour.

A contour plot is appropriate if you want to see how value Z changes as a function of two inputs X and Y such that `X = f(X, Y)`.

The independent variables x and y are usually restricted to a regular grid called __meshgrid__. The numpy.meshgrid creates a rectangular grid out of an array of x values and an array of y values.

Matplotlib API contains **`contour()`** and **`contourf()`** functions that draw contour lines and filled contours, respectively. Both functions need three parameters x,y and z.

```python
xlist = np.linspace(-4.0, 4.0, 100)
ylist = np.linspace(-4.0, 4.0, 100)
X, Y = np.meshgrid(xlist, ylist)
Z = np.sqrt(X**2 + Y**2)
fig,ax = plt.subplots(1,1, figsize=(12, 10))
cp = ax.contourf(X, Y, Z)
x = np.arange(-np.pi, np.pi, 0.01)
y = np.sin(x)
ax.plot(x, y, color='r')
fig.colorbar(cp) # Add a colorbar to a plot
ax.set_title('Filled Contours Plot')
ax.set_xlabel('x (cm)')
ax.set_ylabel('y (cm)')
```

## 7. Box Plot

This plot is used to plot the set of data containing the minimum,first quartile, median, third quartile, and maximum.

```python
np.random.seed(10)
collectn_1 = np.random.normal(100, 10, 200)
collectn_2 = np.random.normal(80, 30, 200)
collectn_3 = np.random.normal(90, 20, 200)
collectn_4 = np.random.normal(70, 25, 200)
# print(collectn_1[:10])

fig = plt.figure()
# Create an axes instance
ax = fig.add_axes([0,0,1,1])
# Create the boxplot
data = [collectn_1, collectn_2, collectn_3, collectn_4]
bp = ax.boxplot(data)
```
